### Using linked list to handle list and find the position of list

```python
# Build node class

class Node():
    def __init__(self, value=None):
        self.value = value
        self.next = None
        self.prev = None

    def __repr__(self):
        return str(self.prev.value if self.prev else None) + " -> [" + str(self.value) + "] -> " + str(self.next.value if self.next else None)

node1 = Node(1)

node1
```

```python
pre_token_nodes = pre_token.encode("utf-8")

pre_token_nodes = [Node(t) for t in pre_token_nodes]

for node, next_node in zip(pre_token_nodes, pre_token_nodes[1:]):
    node.next = next_node
    next_node.prev = node
pre_token_nodes[0:3]
```

```python
counter = dict()
counter_tracking = dict()
for node in pre_token_nodes[:-1]:
    k = (node.value, node.next.value)
    if k in counter:
        counter[k] += 1
        counter_tracking[k].append(node)
    else:
        counter[k] = 1
        counter_tracking[k] = [node]


counter,
#  counter_tracking
```
